```python
import os
import requests as req
import pandas as pd
import numpy as np
```

```python
IRIS_FILE = req.get('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data')
with open(os.getcwd() + '/iris.data', 'w') as fw_iris:
    fw_iris.write(IRIS_FILE.text)
```

```python
IRIS_DF = pd.read_csv(os.getcwd() + '/iris.data', names=['sepal length', 'sepal width', 'petal length', 'petal width', 'class'])
```

# map Method

```python
IRIS_DF['class'] = IRIS_DF['class'].map({'Iris-setosa': 'SET', 'Iris-virginica':
'VIR', 'Iris-versicolor': 'VER'})
IRIS_DF
```

# apply Method

```python
IRIS_DF['wide petal'] = IRIS_DF['petal width'].apply(lambda v: 1 if v >= 1.3 else 0)
IRIS_DF
```

```python
IRIS_DF['petal area'] = IRIS_DF.apply(lambda r: r['petal length'] * r['petal width'],
axis=1)
IRIS_DF
```

# applymap Method

```python
IRIS_DF.applymap(lambda v: np.log(v) if isinstance(v, float) else v)
IRIS_DF
```

# groupby Method

```python
IRIS_DF.groupby('class')['petal width']\
    .agg({'delta': lambda x: x.max() - x.min(), 'max': np.max, 'min': np.min})
```
